```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
path = '/content/drive/MyDrive/haberman.csv'
df = pd.read_csv(path)
df
```

```python
df.columns = ['patient_age', 'operation_year', 'positive_axillary_nodes', 'survival_status']
df
```

```python
df['survival_status'].value_counts()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),  yticklabels = False)
```

```python
df['survival_status'] = df['survival_status'].map({1:"yes", 2:"no"})
```

```python
df
```

```python
df.describe()
```

```python
survival_yes = df[df['survival_status'] == 'yes']
survival_yes.describe()
```

```python
survival_no = df[df['survival_status'] == 'no']
survival_no.describe()
```

```python
sns.FacetGrid(df, hue = "survival_status").map(sns.distplot, "patient_age").add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue = "survival_status").map(sns.distplot, "operation_year").add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue = "survival_status").map(sns.distplot, "positive_axillary_nodes").add_legend()
plt.show()
```

```python
plt.figure(figsize = (15, 4))
plt.subplot(1,3,1)
sns.boxplot(x = 'survival_status', y = 'patient_age', data = df)
plt.subplot(1,3,2)
sns.boxplot(x = 'survival_status', y = 'operation_year', data = df)
plt.subplot(1,3,3)
sns.boxplot(x = 'survival_status', y = 'positive_axillary_nodes', data = df)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue = 'survival_status')
plt.show()
```

```python
sns.jointplot(x = 'patient_age', y = 'positive_axillary_nodes', data = df)
plt.show()
```

```python
sns.heatmap(df.corr(), cmap ='YlGnBu', annot = True)
plt.show()
```

```python
sns.jointplot(x = 'patient_age',  y = 'operation_year' , data = df,  kind = 'kde', fill = True)
plt.show()
```

```python
sns.jointplot(x = 'patient_age',  y = 'positive_axillary_nodes' , data = df,  kind = 'kde', fill = True)
plt.show()
```

```python
sns.jointplot(x = 'operation_year',  y = 'positive_axillary_nodes' , data = df,  kind = 'kde', fill = True)
plt.show()
```

inferences:
1. patients with age 35 or less are more likely to survive for more than 5 years; whereas patients with age more than 70 are less likely to survive for more than 5 years.
2. patients with positive axillary nodes between 0-5 are more likey to survive than patients with positive axillary nodes more than 5.
