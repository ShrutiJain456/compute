```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/Ad Click Data.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
count = 0
for i in df.isnull().sum(axis = 1):
  if i > 0:
    count = count + 1
print('Total number of rows with NaN values:', count)
percent = (count/len(df))*100
print("percent of missing values to entire dataset:", percent)
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
obj_var = ['Ad Topic Line', 'City', 'Country']  
df[obj_var].describe()
```

```python
pd.crosstab(index = df['Country'], columns = 'count').sort_values(['count'],ascending = False).head(15)
```

```python
df = df.drop(['Ad Topic Line', 'City', 'Country'], axis = 1)
df
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df
```

```python
df['Timestamp'].describe()
```

```python
df = df.drop(['Timestamp'], axis = 1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df[['Daily Time Spent on Site',	'Age','Area Income',	'Daily Internet Usage',	'Male']]
y = df['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```
